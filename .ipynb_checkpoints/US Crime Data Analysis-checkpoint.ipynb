```python
import re

def extract_city(city_raw):
    rgx = r'([\w\s]+)\s[Cc]ity$'
    ret_str = ''
    if pd.notnull(city_raw):
        try:
            ret_str = re.search(rgx, city_raw).group(1)
        except AttributeError:
            ret_str = city_raw

        return ret_str.strip().title().replace('Of', 'of')
    else:
        return city_raw
```

```python

week_type = pd.api.types.CategoricalDtype(categories=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], ordered=True)
month_type = pd.api.types.CategoricalDtype(categories=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'], ordered=True)
hour_type = pd.api.types.CategoricalDtype(categories=['12 AM', '01 AM', '02 AM', '03 AM', '04 AM', '05 AM', '06 AM', '07 AM',
                                                      '08 AM', '09 AM', '10 AM', '11 AM', '12 PM', '01 PM', '02 PM', '03 PM',
                                                      '04 PM', '05 PM', '06 PM', '07 PM', '08 PM', '09 PM', '10 PM', '11 PM'], ordered=True)


def extract_date_info(df):
    """
        Extracts time information from a DateTime Series
    """
    df['month'] = df['Date'].dt.month_name()
    df['month'] = df['month'].astype(month_type)
    df['weekday'] = df['Date'].dt.day_name()
    df['weekday'] = df['weekday'].astype(week_type)
    df['quarter'] = df['Date'].dt.quarter
    df['hour'] = df['Date'].apply(lambda x: x.strftime("%I %p")).astype(hour_type)
```

```python
def extract_web_domain_end(x):
    rgx = r'https{0,1}://(w{0,3}\.|)[\w\.-]+(\.\w+)/+?'
    return re.search(rgx, x).group(2)
```

```python
def extract_news_provider(x):
    rgx = r'https{0,1}://(w{0,3}\.|)([\w\.-]+)(\.(biz|blog|ca|com|de|edu|eu|fm|gov|in|info|is|me|mn|net|news|nyc|org|today|tv|uk|us))/+?'
    return re.search(rgx, x).group(2)
```

#### Question 1: Which organizations and providers report crime news most frequently?

```python
def plot_most_frequent(series, title, xlabel, start_ind=0, col_ind=0, kind='bar', ascending=True, x_ticks_rotation=15):
    fig, ax = plt.subplots(figsize=(20,15))
    series[start_ind:10].sort_values(ascending=ascending).plot(kind=kind, color=sns.color_palette()[col_ind], ax=ax)
    plt.xticks(rotation=x_ticks_rotation, fontsize=25)
    plt.yticks(fontsize=25)
    ax.set_title(title, fontsize=40)
    ax.set_xlabel(xlabel, fontsize=30)
    ax.set_ylabel('Frequency', fontsize=30)
    plt.show()
```

```python
org_occurences = crime_df['Organization'].value_counts()
```

```python
plot_most_frequent(org_occurences, 'Most Frequent News Reporting Organizations', 'Organizations', kind='barh', x_ticks_rotation=0)
```

The plot above shows that HuffPost reported the most crime news in 2017

```python
prov_occurences = crime_df['news provider'].value_counts()
```

```python
plot_most_frequent(prov_occurences, 'Most Frequent News Providers', 'Providers', col_ind=1, ascending=False)
```

From the plot above, the most frequent news provider is huffingtonpost

```python
def plot_crime_count(x, col_ind, x_ticks_rotation=15):
    fig, ax = plt.subplots(figsize=(20,8))
    sns.countplot(data=crime_df, x=x, color=sns.color_palette()[col_ind], ax=ax)
    plt.xticks(rotation=x_ticks_rotation, fontsize=20)
    plt.yticks(fontsize=20)
    plt.xlabel(f'{x.capitalize()}', fontsize=20)
    plt.ylabel('Count', fontsize=20)
    plt.title(f'Crime Count by {x.capitalize()}', fontsize=25)
    plt.show()
    
def plot_crime_line(series, col_ind, x, x_ticks_rotation=15):
    fig, ax = plt.subplots(figsize=(20,8))
    sns.lineplot(x=series.index, y=series.values, color=sns.color_palette()[col_ind], ax=ax)
    plt.xticks(rotation=x_ticks_rotation, fontsize=20)
    plt.yticks(fontsize=20)
    plt.xlabel(f'{x.capitalize()}', fontsize=20)
    plt.ylabel('Frequency', fontsize=20)
    plt.title(f'Crime Frequency by {x.capitalize()}', fontsize=25)
    plt.show()
```

```python
plot_crime_count('month',4)
```

```python
crime_count_by_month = crime_df['month'].value_counts()

plot_crime_line(crime_count_by_month, 4, 'month')
```

From the plots above, we see the trend in crimes committed in 2017. No crime was committed in January, the trend of crimes committed seems to be very low towards the start or end of the year and high around the second and third quarter of the year

#### Question 8: What words appear most frequently in the crime news Keywords?

```python
from collections import defaultdict


keyword_freq = defaultdict(int)

for list_of_keywords in crime_df['list of keywords'].values:
    for value in list_of_keywords:
        if value != 'Unknown':
            keyword_freq[value] += 1

keyword_freq = dict(sorted(keyword_freq.items(), key=lambda item: item[1], reverse=True))
```

```python
fig, ax = plt.subplots(figsize=(20,8))
plt.barh(y=list(keyword_freq.keys())[:20][::-1], width=list(keyword_freq.values())[:20][::-1])
plt.xticks(fontsize=15)
plt.yticks(fontsize=20)
plt.xlabel('Frequency', fontsize=20)
plt.ylabel('Keywords', fontsize=20)
plt.title('Most Frequent Keywords', fontsize=25)
plt.show()
```

The most frequent keyword appearing in crime news is **"hate"**
